```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/DELL/Desktop/data1.csv")
```

```python
data = data[::4]
```

```python
a = np.asarray(data['XY'])
```

```python
a
```

```python
spot_dict = {}
spot_pos = []


for i in range (0,100):
        b = a[i]
        c = b.split(',')
        for j in range (0,4):
            d = c[j]
            e = int(d)
            if j==0:
                x1 = e
            elif j==1:
                y1 = e
            elif j==2:
                x2 = e
            elif j==3:
                y2 = e
        cur_len = i
        spot_dict[(x1, y1, x2, y2)] = cur_len + 1
        spot_pos.append(spot_dict)
        
```
